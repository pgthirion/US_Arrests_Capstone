# Capstone Project III: Data Visualization

### Importing the libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Loading of the Dataset

```python
# Load the dataset
data = pd.read_csv('UsArrests.csv')
```

### Cleaning of the Data

```python
# Data Cleaning
data.dropna(inplace=True)  # Remove incomplete entries
```

### Data Stories and Visualizations

#### Bar Graph of Assault Rates by City

```python
# Bar Graph of Assault Rates by City
plt.figure(figsize=(12, 8))
sns.barplot(data=data, x='City', y='Assault')
plt.title('Bar Graph of Assault Rates by City')
plt.xlabel('City')
plt.ylabel('Assault Rate')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig('bar_assault_rates.png')
plt.show()
```

#### Scatter Plot of Rape vs. Murder

```python
# Scatter Plot of Rape vs. Murder
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Rape', y='Murder')
plt.title('Scatter Plot of Rape vs. Murder')
plt.xlabel('Rape')
plt.ylabel('Murder')
plt.tight_layout()  # Adjust layout for optimal spacing
plt.savefig('scatter_rape_vs_murder.png')
plt.show()
```

#### Pairplot of Numerical Variables

```python
# Pairplot of Numerical Variables
sns.pairplot(data=data)
plt.suptitle('Pairplot of Numerical Variables', y=1.02)  # Title positioning
plt.tight_layout()  # Adjust layout for optimal spacing
plt.savefig('pairplot_numerical_variables.png')
plt.show()
```

#### Line Plot of UrbanPop and Murder

```python
# Line Plot of UrbanPop and Murder
plt.figure(figsize=(10, 8))
sns.lineplot(data=data, x='UrbanPop', y='Murder')
plt.title('Line Plot of UrbanPop and Murder')
plt.xlabel('UrbanPop')
plt.ylabel('Murder')
plt.tight_layout()  # Adjust layout for optimal spacing
plt.savefig('line_urbanpop_vs_murder.png')
plt.show()
```

#### Boxen Plot of Rape by UrbanPop

```python
# Boxen Plot of Rape by UrbanPop
plt.figure(figsize=(10, 6))
sns.boxenplot(data=data, x='UrbanPop', y='Rape')
plt.title('Boxen Plot of Rape by UrbanPop')
plt.xlabel('UrbanPop')
plt.ylabel('Rape')
plt.tight_layout()  # Adjust layout for optimal spacing
plt.savefig('boxen_rape_by_urbanpop.png')
plt.show()
```

#### Bar Graph of Assault by UrbanPop

```python
# Bar Graph of Assault by UrbanPop
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='UrbanPop', y='Assault')
plt.title('Bar Graph of Assault by UrbanPop')
plt.xlabel('UrbanPop')
plt.ylabel('Assault')
plt.tight_layout()  # Adjust layout for optimal spacing
plt.savefig('bar_assault_by_urbanpop.png')
plt.show()
```
